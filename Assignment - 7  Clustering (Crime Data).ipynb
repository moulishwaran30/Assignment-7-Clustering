# Hierarchical clustering

```python
def norm_func(i):
    x = (i - i.min())/(i.max() - i.min())
    return (x)
```

```python
hier_crime = norm_func(crime_data.iloc[:,1:])
hier_crime
```

```python
plt.figure(figsize=(15,15))
hier_crime1 = sch.dendrogram(sch.linkage(hier_crime , method = 'single'))
```

```python
hier_crime2 = AgglomerativeClustering(n_clusters = 4 ,affinity = 'euclidean', linkage = 'single')
hier_crime2
```

```python
hier_crime3 = hier_crime2.fit_predict(hier_crime)
hier_crime3
```

```python
hier_crime4  = pd.DataFrame(hier_crime3,columns=['clustering'])
hier_crime4
```

```python
crime_data1 = copy.copy(crime_data)
crime_data1
```

```python
crime_data1['clustering'] = hier_crime4
crime_data1
```

```python
crime_data1[crime_data1['clustering'] == 0]
```

```python
crime_data1[crime_data1['clustering'] == 1]
```

```python
crime_data1[crime_data1['clustering'] == 2]
```

```python
crime_data1[crime_data1['clustering'] == 3]
```

# K - Means clustering

```python
scaler = StandardScaler()
kmean = scaler.fit_transform(crime_data.iloc[:,1:])
print(kmean)
```

```python
wcss = []
for i in range (1,11):
    kmean1 = KMeans(n_clusters=i,random_state=0)
    kmean1.fit(kmean)
    wcss.append(kmean1.inertia_) 
```

```python
wcss
```

```python
kmean2 = KMeans(n_clusters=4,random_state=12)
kmean2.fit(kmean)
```

```python
kmean2.labels_
```

```python
crime_data2 = copy.copy(crime_data)
crime_data2
```

```python
crime_data2['clustering'] = kmean2.labels_
```

```python
crime_data2
```
